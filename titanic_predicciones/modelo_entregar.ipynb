```python
import pandas as pd
```

```python
## primero hay que cargar las 4 bases de datos
```

```python
x_train = pd.read_csv('bases_procesadas/x_train.csv', index_col=0)
x_train.head()
```

```python
y_train = pd.read_csv('bases_procesadas/y_train.csv', index_col=0)
y_train.head()
```

```python
x_test = pd.read_csv('bases_procesadas/x_test.csv', index_col=0)
x_test.head()
```

```python
y_test = pd.read_csv('bases_procesadas/y_test.csv', index_col=0)
y_test.head()
```

```python
## usaremos el random forest classifier
```

```python
## primero tuneamos hyperparametros
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
```

```python
rf = RandomForestClassifier()
```

```python
param_grid =[{'n_estimators':[10,20,50,70,100],'max_depth':[3,5,6,7,8,9,10,12,15],'criterion':['gini','entropy']}]
```

```python
clf = GridSearchCV(rf, param_grid=param_grid, cv=5)
```

```python
best_clf = clf.fit(x_train, y_train)
```

```python
best_clf.best_estimator_
```

```python
print (f'Accuracy - : {best_clf.score(x_train,y_train):.5f}')
```

## Montamos el Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
rf = RandomForestClassifier(criterion='entropy', max_depth=5, n_estimators=10)
```

```python
rf.fit(x_train, y_train)
y_pred = rf.predict(x_test)
score = round(rf.score(x_train, y_train)*100 , 2)

print(score)
```

## Guardamos las predicciones en un .csv nuevo

```python
submission_df = pd.DataFrame(columns=['PassengerId', 'Survived'])

submission_df['PassengerId'] = y_test['PassengerId']

submission_df['Survived'] = y_pred

submission_df.to_csv('submissions_8.csv', header=True, index=False)
submission_df.head(10)
```
