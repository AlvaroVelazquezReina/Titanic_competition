## Tratamiento a train:

```python
train.info()
```

```python
## quitamos columnas que no interesan
train = train.drop(['PassengerId', 'Ticket', 'Name', 'Cabin'], axis=1)
train.head()
```

```python
## quitamos los datos nulos, podriamos hacerle una media a las edades o algo asi, pero mejor lo dropeamos
train.isnull().sum()
```

```python
train = train.dropna()
```

```python
train.isnull().sum()
```

```python
train.shape
```

```python
## label encoding a 'sex'
from sklearn import preprocessing

label_encoding = preprocessing.LabelEncoder()
train["Sex"] = label_encoding.fit_transform(train["Sex"].astype(str))

train.head()
```

```python
## ahora label encoding a 'embarked'

label_encoding = preprocessing.LabelEncoder()
train["Embarked"] = label_encoding.fit_transform(train["Embarked"].astype(str))

train.head()
```

```python
## ahora creamos de SibSp + Parch  = Fsize
## y le hacemos el one_hot_encoding
```

```python
## creamos Fsize
sum_column = train["SibSp"] + train["Parch"]
train['Fsize'] = sum_column
train.head()
## vale tenemos SibSp y Parch juntas en Fsize
```

```python
# dropeamos SibSp y Parch
train = train.drop(['SibSp', 'Parch'], axis=1)
train.head()
```

```python
## ahora partimos en 5 ese Fsize y le hacemos one hot encoding
train['Alone'] = train['Fsize'].map(lambda s: 1 if s == 0 else 0)
train['Single'] = train['Fsize'].map(lambda s: 1 if s == 1 else 0)
train['SmallF'] = train['Fsize'].map(lambda s: 1 if  s == 2  else 0)
train['MedF'] = train['Fsize'].map(lambda s: 1 if 3 <= s <= 4 else 0)
train['LargeF'] = train['Fsize'].map(lambda s: 1 if s >= 5 else 0)
```

```python
train.head()
```

```python
##dropeamos Fsize del dataset
train = train.drop(['Fsize'], axis=1)
train.head()
```

```python
train.shape
```

```python
train.columns
```

```python
## y ahora la partimos en x_train e y_train
x_train = train.drop(['Survived'],axis=1)
x_train.head()
```

```python
x_train.shape
```

```python
y_train = train.drop(['Pclass', 'Sex', 'Age', 'Fare', 'Embarked', 'Alone',
       'Single', 'SmallF', 'MedF', 'LargeF'], axis=1)
y_train.head()
```

```python
y_train.shape
```

```python
## y guardamos las dos datasets creadas por nosotros
```

```python
## guardamos x_train
x_train = pd.DataFrame(x_train)
x_train.to_csv('bases_procesadas/x_train.csv')
```

```python
## guardamos y_train
y_train = pd.DataFrame(y_train)
y_train.to_csv('bases_procesadas/y_train.csv')
```
